```python
import mpld3
from mpld3 import plugins
from mpld3.utils import get_id
import numpy as np
import collections
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(figsize=(12, 8))
x = np.arange(len(sub_data.location.unique()))

# Define bar width. We'll use this to offset the second bar.
bar_width = 0.4

# Note we add the `width` parameter now which sets the width of each bar.
b1 = ax.bar(x, sub_data.loc[sub_data['month'] == 'April', 'mean_total_cases_per_million'],
            width=bar_width, label = 'April')

b2 = ax.bar(x + bar_width, sub_data.loc[sub_data['month'] == 'September', 'mean_total_cases_per_million'],
            width=bar_width, label = 'September')

ax.set_xticks(x + bar_width / 2)
ax.set_xticklabels(sub_data.location.unique())

# Use Seaborn's context settings to make fonts larger.
import seaborn as sns
sns.set_context('talk')

# Add legend.
ax.legend()

# Axis styling.
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('#DDDDDD')
ax.tick_params(bottom=False, left=False)
ax.set_axisbelow(True)
ax.yaxis.grid(False)
ax.xaxis.grid(False)

# Add axis and chart labels.
ax.set_xlabel('Country', labelpad=15)
ax.set_ylabel('Total Cases per Million', labelpad=15)
ax.set_title('Total covid19 cases per million by month', pad=15)

fig.tight_layout()

mpld3.display()
```
